```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from scipy import stats
```

```python
# Generate some sample data
np.random.seed(0)
scores = np.random.randint(50, 100, size=(100, 3))  # 100 students, 3 exams
```

```python
# Create a Pandas DataFrame from the NumPy array
df = pd.DataFrame(scores, columns=['Exam 1', 'Exam 2', 'Exam 3'])
```

```python
# Calculate summary statistics using NumPy and Pandas
mean_scores = np.mean(scores, axis=0)
median_scores = np.median(scores, axis=0)
std_scores = np.std(scores, axis=0)
```

```python
df_summary = pd.DataFrame({
    'Mean': mean_scores,
    'Median': median_scores,
    'Standard Deviation': std_scores
}, index=df.columns)
```

```python
print("Summary Statistics:\n", df_summary)
```

```python
# Visualize the data using Matplotlib
plt.figure(figsize=(8, 6))
plt.boxplot(scores, labels=df.columns)
plt.title('Exam Scores Distribution')
plt.xlabel('Exams')
plt.ylabel('Scores')
plt.show()
```

```python
# Perform a statistical test using Scipy
t_statistic, p_value = stats.ttest_ind(scores[:, 0], scores[:, 1])
print(f"\nT-Test Results (Exam 1 vs. Exam 2):\n"
      f"T-Statistic: {t_statistic:.2f}, P-Value: {p_value:.4f}")
```
